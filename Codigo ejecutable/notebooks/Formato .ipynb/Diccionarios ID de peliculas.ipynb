### Diccionarios ID de películas
En este notebook se crean diccionarios que permiten al sistema obtener el título de la película a partir de us identificador, obtener el identificador a partir del título y la posición que ocupa la película en nuestro dataset a partir de su ID. El sistema necesita accesos rápidos a esta información y por eso precalculamos estos diccionarios en lugar de recuperarla durante la ejecución.

Importamos las librerías necesarias

```python
import pandas as pd
import pickle
```

Cargamos y guardamos en el fichero movieIDTitleTable.pkl un diccionario de ID de película a su título

```python
df_movies_catalog = pd.read_csv("movies_catalog_clean.csv")
movieIDTitle = df_movies_catalog[['id', 'title']].set_index('id').to_dict()['title']
print(movieIDTitle)
fileTable = open("movieIDTitleTable.pkl", "wb")
pickle.dump(movieIDTitle, fileTable)
fileTable.close()
```

Cargamos y guardamos en el fichero movieTitleIDTable.pkl el diccionario inverso: de título de película a su ID

```python
df_movies_catalog = pd.read_csv("movies_catalog_clean.csv")
movieTitleID = df_movies_catalog[['id', 'title']].set_index('title').to_dict()['id']
fileTable = open("movieTitleIDTable.pkl", "wb")
pickle.dump(movieTitleID, fileTable)
fileTable.close()
```

Dado un ID de película, en el agente recomendador necesitaremos acceso constante a la posición que ocupa en las tablas (en particular qué fila del Numpy Array que constituye el modelo content based es la de sus recomendaciones). Creamos un diccionario que recoja esta información

```python
movieIDIndex = dict(zip(df_movies_catalog.id, df_movies_catalog.index))
print(movieIDIndex)
fileTable = open("movieIDIndexTable.pkl", "wb")
pickle.dump(movieIDIndex, fileTable)
fileTable.close()
```
